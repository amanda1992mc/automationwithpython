## Desafio:
### Enviar um e-mail com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

Comandos pyautogui: https://pyautogui.readthedocs.io/en/latest/quickstart.html

```python
#Passos:

#1. Entrar no sistema/pasta/link e fazer o download do arquivo
#2. Abrir o arquivo e fazer os cálculos necessários
#3. Abrir o email e fazer o envio dos dados
```

```python
import pyautogui
import pyperclip
import time
import pandas as pd
```

### 1. Acessando do arquivo:

```python
#acessar link do arquivo

pyautogui.pause = 1
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/14oLE59U1RqyRqlBbKpsyymW-mitvbtoh")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(5)

pyautogui.click(x=490, y=406)
time.sleep(1)
pyautogui.click(x=1643, y=281)
time.sleep(1)
pyautogui.click(x=1423, y=873)
time.sleep(3)
pyautogui.press("enter")
time.sleep(3)

```

### 2.Abrir o arquivo e fazer os cálculos

```python
tabela = pd.read_excel(r"Vendas - Dez.xlsx")
display(tabela)
```

```python
#quantidade = soma da coluna quantidade
quantidade = tabela['Quantidade'].sum()

#faturamento = soma da coluna valor final
faturamento = tabela["Valor Final"].sum()
```

### 3. Abrir o email e fazer o envio dos dados

```python
pyautogui.pause = 1
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(5)
pyautogui.click(x=121, y=298)
time.sleep(5)
#pyperclip.copy("destinatario@gmail.com")
pyautogui.write("destinatario@gmail.com")
pyautogui.press("tab")
pyautogui.press("tab")
date = time.strftime('%d/%m/%Y')
pyautogui.write(f"Vendas e Faturamento de {date}")
pyautogui.press("tab")
time.sleep(2)

texto = f"""
Prezados,

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abs
"""
pyautogui.write(texto)

time.sleep(2)

pyautogui.hotkey("ctrl","enter")
```
